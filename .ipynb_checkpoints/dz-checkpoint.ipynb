```python
import numpy as np
import pandas as pd
import xgboost as xgb
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

from sklearn.linear_model import LogisticRegression
from sklearn import metrics

random_state = 42
```

Lade die Daten von der csv file

```python
df = pd.read_csv("bank+marketing/bank/bank-full.csv", sep = ';')
```

```python
df.head()
```

Manche Variablen beinhalten Wörter
Der BDT braucht aber Zahlen zum Schneiden
Für den BDT sollte die Hirachie der Wörter mehr oder weniger egal sein
Besonders in der Bildung und Job könnte man eine Hirachie erstellen, es sollte das Ergebnis etwas verbessern, aber in den 2 Stunden kam ich nicht mehr dazu. Würde ich mit mehr Zeit angehen.
Auf jedenfall werden die Wörter durch Zahlen ersetzt

```python
replaceList = ['job', 'marital', 'education', 'default', 'balance','housing', 'contact', 'loan', 'duration', 'poutcome', 'y', 'month']

for name in replaceList:
    ipWord = df[name].unique()
    word_dict = dict(zip(ipWord, range(len(ipWord))))
    for key in word_dict.keys():
        df[name] = df[name].replace(key, word_dict[key])
```

```python
df['employed']=0
```

Arbeitslos ist Nummer 9

Erzeuge eine neue Column, vielleicht hilft es

```python
df['employed'].loc[df['job'] != 9 ] =1
df['employed'].loc[df['job'] == 9 ] =0
df.insert(len(df.columns)-1, 'y', df.pop('y'))
```

Selektiere alle Variablen außer der Target Variable

```python
trainVariables = df.columns
trainVariables.drop('y')
```

```python
df.head()
```

Es wird ein Trainingsdatensatz und ein Validierungsdatensatz erzeugt
BDTs neigen sehr stark zum Overfitten
Es führt kein Weg am Validierungsdatensatz vorbei
Idealerweise hätte man noch einen Testing Datensatz, mit dem am Ende alles evaluiert wird. (Dieser Datensatz wird nicht beim Trainning in den BDT gesteckt)

```python
#Training/Validation Split
X_train, X_val, y_train, y_val = train_test_split(df, df['y'], test_size=0.30, random_state = random_state)
```

Setze den BDT auf
Errfahrungsgemäß performen BDT auf tabellierten Daten vergleichbar zu Neuronalen Netzwerken und BDTs sind deutlich leichter zu handhaben. (Für NN müsste die Hirachie bereinigt werden) 
Daher BDT.

```python
X_train_NoY = X_train.drop(['y'],axis = 1)
X_val_NoY = X_val.drop(['y'],axis = 1)


param_classifier_single = {
    'tree_method': 'exact',
    'njobs' : 4,
    'learning_rate': 0.06,
    'max_depth':4,
    'n_estimators': 10000,
    'colsample_bynode':0.9,
    'subsample':0.9,
    'colsample_bytree':0.9,
    'booster':'gbtree'
}

model = xgb.XGBClassifier(
    objective = 'binary:logistic',
    **param_classifier_single
    )
```

Trainiere den BDT
Der BDT versucht die AUC (Area under Curve ROC) zu maximieren

```python
h = model.fit(          X_train_NoY, y_train,
                        eval_metric=['auc'],
                        eval_set=[(X_train_NoY, y_train),
                                 (X_val_NoY, y_val)],
                        early_stopping_rounds = 100,
                            )
```

Es ist immer interessant zu verstehen, welche Variablen für den BDT besonders wichtig sind
-> Feature Importance

```python
sorted_idx = np.argsort(model.feature_importances_)[::-1]
index_list = []
importance_list = []
for index in sorted_idx:
    index_list.append(trainVariables[index])
    importance_list.append(model.feature_importances_[index])
df_imp  = pd.DataFrame({'Feature':index_list, "Importance":importance_list})
```

poutcome ist die wichtigste Variable:
Ergibt auch irgendwo Sinn, da wenn man vorher schon einen Vertrag abgeschlossen hat, dann ist die wahrscheinlichkeit höher nochmal einen abzuschließen, würde ich mir jetzt denken
interessanter sind die anderen Variablen
contact ist die zweit wichtigste, obwohl es hier nur die Möglichkeit Festnetz oder Handy gab. Da es sich um einen recht alten Datensatz (2008-2010) handelt, vermute ich, dass es damals noch nicht so viele Handys gab und es eher eine versteckte Korrelation gibt.
Besonders fällt noch auf, dass Jobs angeblich am unwichtigsten ist. Es könnte doch mit der Hirachie zu tun haben.
Interessanterweise ändert es auch nichts, ob die Person generell angestellt ist, oder nicht. 

In weiteren Iterationen könnte man die Performance von diesem Model mit einem vergleichen, bei dem nur die wichtigsten Variablen drin sind. Allerdings müssten vorher die genauen Korrelationen verstanden werden.


```python
df_imp
```

```python
X_val['pred'] = model.predict_proba(X_val_NoY)[:,1]
```

Seperation Plot
y=no wird ziemlich verlässlich vorhergesagt.
Man beachte die logarythmische Achse
Welche Leute man allerdings nicht mehr Kontaktiert, in der nächsten Kampagne kann man aus diesen Plots nicht ablesen.
Dafür müsste man den Gewinn eines Vertragsabschluss und die Kosten der Kampagne kennen.


Interessant ist auch, dass man häufig ziemlich gut vorhersagen kann, wenn kein Vertrag abgeschlossen wird. Andersrum ist es sehr schwer bzw. das Ergebnis ist mehr oder weniger flach.



```python
plt.hist(X_val.query('y ==0')['pred'], bins = 30, density = False, label = 'y=no')
plt.hist(X_val.query('y ==1')['pred'], bins = 30, alpha= 0.6, density = False, label = 'y=yes')
plt.yscale('log')
plt.xlabel('pred')
plt.title('')
plt.legend()
```

Da der BDT nach der AUC Metric einer ROC Kurve trainiert wurde, ist es auch interessant, sich die ROC Kurve anzuschauen.
Dabei erkennt man schon, dass der trainierte BDT deutlich besser als ein untrainiertes Model, welches random entscheidet.
Die AUC des BDT ist bei 0.87, was deutlich über 0.5 (No Skill) ist

```python
log_regression = LogisticRegression()

log_regression.fit(X_train,y_train)

y_pred_proba = log_regression.predict_proba(X_val.drop(['y'],axis =1))[::,1]
fpr, tpr, _ = metrics.roc_curve(y_val,  y_pred_proba)

#create ROC curve
plt.plot(fpr,tpr, label = 'Validation Dataset')
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')


plt.plot([0,1],  [0,1], label = 'No Skill', color = 'k')
plt.legend()
plt.title('ROC')
plt.show()
```

Ich habe mich für ROC entschieden, da die Charakteristik generell sehr stabil ist.
Figure of Merit empfand ich immer als zu instabil und man hat aus der Kurve an sich nicht so viel Spielraum. Es peakt an einer Stelle und aus dem Rest kann man nicht viel ablesen.

Beim Zusammenschreiben ist mir noch eingefallen, dass es für Datensätze in denen Signal (in unserem Fall y=yes) unterrepräsentiert ist, die Precision Recall Curve in der Regel besser als die ROC Curve ist. Wenn man das Projekt weiter ausführen würde, dann wäre das auch eine Änderung, welche ich auf jeden Fall im Hinterkopf behalten würde.
